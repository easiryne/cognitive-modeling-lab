```python
import numpy as np
```

```python
evidence = 0.1
stimulus = -0.2
```

```python
if abs(stimulus) > abs(evidence):
    evidence = stimulus
```

```python
evidence = np.array([
    [0.],
    [0.1],
    [-0.1],
])
stimulus = np.array([
    [0.1],
    [0],
    [0.2]
])
```

```python
condition = np.abs(stimulus) > np.abs(evidence)
condition
```

```python
evidence = np.where(condition, stimulus, evidence)
```
